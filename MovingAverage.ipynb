```
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import signal
```

```
def movingAverage(x,M1,M2):
  y = np.zeros(len(x)) 
  for n in range(0,len(y)):
    y[n]= 1 / (M1 + M2 +1 ) * np.sum(x[n:n+M1+M2+1])
  return y
```

```
n = np.linspace(-50,49,100, dtype=int)
x = np.random.randint(1,10,100)
y = movingAverage(x,0,3)
```

```
plt.stem(n[:50],x[:50],label='x')
plt.stem(n[:50],y[:50],'r',label='y')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.figure()
plt.plot(n[:50],x[:50],marker='*',label='x')
plt.plot(n[:50],y[:50],'r',marker='*',label='y')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
